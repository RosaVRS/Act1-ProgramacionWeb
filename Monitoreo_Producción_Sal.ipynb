```python
import matplotlib.pyplot as plt
import pandas as pd

# Función para agregar datos
def agregar_datos(dia, temperatura, salinidad, velocidad_viento, cantidad_sal):
    datos.loc[dia] = [temperatura, salinidad, velocidad_viento, cantidad_sal]

# Inicializar DataFrame
datos = pd.DataFrame(columns=['Temperatura (°C)', 'Salinidad (g/L)', 'Velocidad del Viento (km/h)', 'Cantidad de Sal (toneladas)'])

# Ejemplo de ingreso de datos para 30 días
for dia in range(1, 31):
    agregar_datos(f'Día {dia}',
                   temperatura=20 + (dia % 10),  # Simulación de temperatura
                   salinidad=35,  # Salinidad constante
                   velocidad_viento=5 + (dia % 5),  # Simulación de velocidad del viento
                   cantidad_sal=100 + (dia * 2))  # Producción de sal

# Graficar los datos
plt.figure(figsize=(15, 5))

# Gráfico de producción de sal
plt.subplot(1, 3, 1)
plt.plot(datos.index, datos['Cantidad de Sal (toneladas)'], label='Producción de Sal', color='blue')
plt.title('Producción de Sal a lo Largo de 30 Días')
plt.xlabel('Día')
plt.ylabel('Toneladas')
plt.legend()

# Graficar temperatura
plt.subplot(1, 3, 2)
plt.plot(datos.index, datos['Temperatura (°C)'], label='Temperatura (°C)', color='red')
plt.title('Temperatura a lo Largo de 30 Días')
plt.xlabel('Día')
plt.ylabel('°C')
plt.legend()

# Graficar salinidad
plt.subplot(1, 3, 3)
plt.plot(datos.index, datos['Salinidad (g/L)'], label='Salinidad (g/L)', color='green')
plt.title('Salinidad a lo Largo de 30 Días')
plt.xlabel('Día')
plt.ylabel('g/L')
plt.legend()

plt.tight_layout()
plt.show()
```
